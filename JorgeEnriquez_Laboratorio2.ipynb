```python
#Importar los módulos necesarios para el desarrollo de un modelo de árbol de decisión.

import pandas
from sklearn import tree
import pydotplus
from sklearn.tree import DecisionTreeClassifier
%matplotlib inline
import matplotlib.pyplot as plt
import matplotlib.image as pltimg
```

```python
#Cargar y leer los datos del archivo plano denominado ‘bill_authentication.csv’ utilizando la librería panda.

df = pandas.read_csv('bill_authentication.csv')
print(df)
```

```python
#Elegir características, es decir, columnas de las que intentamos predecir. Se deben elegir las columnas que aportarán más
#al objetivo del modelo.

features = ['Variance', 'Skewness', 'Curtosis', 'Entropy']
X = df[features]
```

```python
#Elegir el objetivo, es decir, la columna con los valores que intentamos predecir.

y = df['Class']
```

```python
#Imprima ‘X’ y ‘y’ para verificar su ecuación.

print(X, y, sep='\n\n')
```

```python
#Crear un árbol de decisiones utilizando DecisionTreeClassifier6 de la librería sklearn. Mediante método “fit” construya un
#clasificador de árbol de decisión a partir del conjunto de entrenamiento (X, y).

dtree = DecisionTreeClassifier()
dtree = dtree.fit(X, y)
```

```python
#Guardar el árbol de decisiones como una imagen utilizando pydotplus.

data = tree.export_graphviz(dtree, out_file=None, feature_names=features)
graph = pydotplus.graph_from_dot_data(data)
graph.write_png('mydecisiontree.png')
```

```python
#Finalmente visualizamos el árbol de decisión utilizando la librería matplotlib

width=1849 # pixeles
height=452
margin=50 # pixeles
dpi=100. # puntos por pulgada

figsize=((width+2*margin)/dpi, (height+2*margin)/dpi) # pulgadas
left = margin/dpi/figsize[0] #proporción de los ejes
bottom = margin/dpi/figsize[1]

fig = plt.figure(figsize=figsize, dpi=dpi)
fig.subplots_adjust(left=left, bottom=bottom, right=1.-left, top=1.-bottom)

img = pltimg.imread('mydecisiontree.png')
plt.imshow(img)
plt.axis('off')
plt.show()
```
